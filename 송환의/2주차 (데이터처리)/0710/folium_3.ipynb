```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
# apply , map 잘써야한다
# applymap                  연습!!
```

```python
tips.sex.map({'Male':1, 'Female':2}) ## 맵은 시리즈에    딕셔너리, 함수써도된다.   ##브로드캐스트
tips.sex.map(lambda x:1 if x=="Male" else 2)  ## 시리즈에 있는 데이터 하나씩 들어옴
        ## 판다스 데이터랭글링 툴, 가져오고, 프로세싱, 보여주고, 간단하게 보여주는거
    ## 맵 
    ## 엑셀에 비해 판다스 쓰는 이유 속도 빠르다
    ## 맵은 시리즈 + 데이터 프레임에 쓸수 있따.
tips = tips.sex.apply(lambda x:1 if x=="Male" else 2) 
```

```python
tips
```

```python
tips.apply(lambda x : x['total_bill']+x['tip'], axis=1)   ## 시리즈에 쓸떄는맵하고 똑가다
```

```python
tips.sex.apply(lambda x:1 if x=="Male" else 2) 
```

```python
tips[['total_bill','tip']].applymap(lambda x:x+1)
```

```python
tips['ratio'] = tips.apply(lambda x:x['tip']/x['total_bill'], axis=1) ## 비율 넣어서 새로운 축 추가
                                                                    ## 고수들 map, apply 잘씀 -> 전처리하다보면 익숙해짐
                                                                    ## 빨라서 효율적, 차수 늘리면
                                                                    ## 차원이 커지면 차원의 저주 그만큼 데이터가 필요 데이터없으면 학습이 안됨
                                                                    ## 학습 안되면 왜곡된 결과 데이터 많으면 많을수록 무조건 좋음
                                                                    ## + transform
```
