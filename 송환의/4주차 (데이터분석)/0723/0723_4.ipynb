```python
tips.boxplot()
```

```python
tips[['total_bill','tip']].boxplot() # 팬시인덱싱
## 이정도 식사금액에 ~~이상주면 아웃라이어이다.
## 예측분석에서는 지원 x but 판다스 aggregation 분석 지원
```

```python
tips.groupby('sex').mean()  ## 비율상 여자가 더 팁 많이줌
```

```python
tips.groupby('sex').sum() # min, max 이거하는게 aggratation 분석
tips.groupby('day').sum() #날짜로 묶어가지고 최소값 최대값~~
```

```python
tips.groupby(['sex','day'])['total_bill', 'tip'].mean() # 3개중에서 보고싶은거 2개다 팬시 인덱싱
```

```python
tips.groupby(['sex','day','time'])['total_bill', 'tip'].mean() # time 어떤상황일떄 팁 제일많이
                                                            # 일요일 저녁에 오는 남자가 팁많이준다
                                                            # 어그리게이션 분석 이렇게 하면 사기꾼 경향보여주기에는 쓰임
```

```python
t = tips.groupby(['sex'])['total_bill', 'tip'].mean() 
```

```python
t.plot.barh() # 판다스에서 하는 분석 기본적인 값 분석 기초통계분석, agrreation 불러서 그림그리는거 끝
```
