```python
from sklearn.neighbors import KNeighborsClassifier # 코딩타입이 클래스
```

```python
knn = KNeighborsClassifier()  ## 하이퍼 파라미터. 파라미터에 따라서 모델 성능이 바뀐다. 디폴트값 바꿔서 모델
                                ## 모델 셀렉션 관점( 최적의 모델 찾는과정)에서 우리는 공부한다.
                                ## 파라메터 바꿔가면서 최적의 알고리즘(모델) 찾는다.
```

```python
# 인공지능 60% 바비블, 피터노빅 //인공지능1 : 현대적 접근방식
# 피터노빅 머신러닝 최고 권위자 왈 기계학습하기위해서는 정규식 중요하다!
# 알고리즘별로 비교 실험 -> 근데 더 중요한건 좋은 데이터의 양 알고리즘 성능 차이 없다.
# 좋은데이터 별로 x -> 그래서 우리는 알고리즘이 중요
# 우리가 모으는 데이터 현실반영 x  따라서 데이터 많으면 많을수록 좋다 얻어걸린다.

##
```

```python
# 훈련, 적합시킨다. fit
# 트레이닝 데이터, 타켓 데이터 분리!
# 행렬 대문자, 벡터 소문자
try:
    knn2 = knn.fit(iris.iloc[:,:-1], iris[iris.columns[-1]])
except ValueError as err:
    pass 
```

```python
## 학습 시키면 예측할수 있따.

#################3 re!!!
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3, n_jobs=-1) ## 기본 k가 5 ## 파라메터 가장좋은거 찾는게 모델 셀렉션 -> 랜더마이징 서칭 해서 쉽게 
                                                    ## -1 자동으로 내컴 최대 효율로 학습시킨다 

knn2 = knn.fit(iris.iloc[:,:-1],iris[iris.columns[-1]])

knn2.predict([[3,3,3,3]])
```

```python
knn2.predict_proba([[3,3,3,3]])  ## 정규화시켜준다. 5개 중 4개는 세토사 하나는 버지니카
```

```python
knn2.predict_proba([[3,3,3,2]])  ## 정규화시켜준다. 5개 모
```
