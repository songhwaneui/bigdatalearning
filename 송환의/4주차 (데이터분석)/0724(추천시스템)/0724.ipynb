```python
import numpy as np
```

```python
t.agg(['count','sum']) ## agg랑 aggregation이랑 같다 agg 하면 표에 한거번에 여러개 보여줄수 있따.
```

```python
t.agg(['count',np.std]) ## 2가지 동시에 표현 // 타이드 형태면 grouby 쑬수있고 기계학습도
t.aggregate(['count',np.std])
```

```python
t.transform("sum")
```

```python
t.transform("count")  ## 그룹바이 했떤걸 각각의 어떤애로 바꿔줌
                        ## 그룹바이 내부구조, sac 기법, 집계함수, aggregate 2개이상, transform 써가지고 각각의 값 바꿀수 있따. 
```
