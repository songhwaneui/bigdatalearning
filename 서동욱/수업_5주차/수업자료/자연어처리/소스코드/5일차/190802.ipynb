```
# Word Synset

import nltk
nltk.download('wordnet')
from nltk.corpus import wordnet
```

```
syn = wordnet.synsets('drink')[0]
print(syn)
print(syn.definition()) # 의미가 출력이 된다.
print(syn.examples()) # 가지고 있는 예문
```

```
syn = wordnet.synsets('motor_vehicle')[0] # 워드넷은 띄어쓰기를 언더바 형식으로 지원함.
print(syn.hypernyms())
print(syn.hypernyms()[0].hyponyms())
print(syn.root_hypernyms()) # entity로 정의를 해서 계층관계로 상위어 하위어로 구성이 되어있다.
```

```
print(syn.pos()) # 가장 많이 사용되는 것을 추출함. (여기서는 명사여서 명사 표시)
print(wordnet.synsets('drink',pos='n')) # 명사로 사용 되는 것만 추출
```

```
syn = wordnet.synsets("drank")[0]
lemmas = syn.lemmas()# 기본형 단어

print(lemmas)
```

```
syn1 = wordnet.synset('good.n.02')
syn2 = syn1.lemmas()[0].antonyms()[0] # 반의어
print(syn2)
```
