```python
def recommendation_user(user_id, k, n=10):
    # 나하고 제일 비슷한 사람 찾음
    s = reco_corr.loc[user_id].sort_values(ascending=False).iloc[1:k+1].index
    temp = pd.DataFrame(columns = ['user_id', 'item_id', 'rating'])
    for j, i in enumerate(s, 1):
        # 나하고 비슷한 사람들이 본 영화 중 내가 안 봤던 영화 찾아냄
        t = set(data[data.user_id == i].item_id) - set(data[data.user_id == user_id].item_id)
        k = data[(data.item_id.isin(t)) & (data.rating==5) & (data.user_id ==i)]
        k.rating = k.rating * (1-0.05*j)
        # 그 중 rating이 5인 애들을 뽑아내줌 (이 부분은 변화할 수 있음. 전략적으로)
        temp = pd.concat([temp, k])
    else:
        # merge하려면 column명이 같아야 함
        temp = temp.rename(columns = {'item_id':'movie_id'})
        return [x for x in temp.merge(item[['movie_id', 'movie_title']]).drop_duplicates().groupby('movie_title').mean().sort_values('rating', ascending=False)[:n].index]
    
    
```
