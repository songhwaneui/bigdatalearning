주의점 : 모두 성능을 바꾸는데 성능의 종류가 여러가지  
우리가 기본적으로 하는 건 정확도를 이야기함  
n_jobs는 n_이 붙어 있으면 Number of의 뜻이 많다고 했음  
초기값에서 n_jobs = None  
n_jobs는 몇 개의 col을 사용해서 학습시킬 것인가?
보통 n_jobs를 하면 학습이 동시에 여러 개의 col을 쓰니까 빨라짐  
모델은 정확도도 중요하지만 예측 속도도 중요  
예측할 때 시간이 많이 걸리면 사용자 입장에서 안 좋음  
딥러닝은 학습은 오래 걸리지만 예측은 빨리 됨  
n_jobs = -1하면 현재 내가 가진 core 대부분을 써서 최대의 리소스를 가지고 학습시킬 수 있다.

```python
knn2 = KNeighborsClassifier(3, n_jobs=-1)
```

```python
knn2.fit(iris.iloc[:,:-1], iris[iris.columns[-1]])
```

```python
knn2.predict_proba([[3,3,3,2]])
```

성능은 정확도뿐 아니라 속도도 포함 <- CS관점
