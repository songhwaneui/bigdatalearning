```python
from sklearn.model_selection import train_test_split # 트레이닝 데이터와 테스트 데이터를 쪼개줌
```

```python
# 판다스만 가지고 쪼갤 수 있음
# iris.sample(
#     n=None,
#     frac=None,
#     replace=False,
#     weights=None,
#     random_state=None,
#     axis=None,
# )
train = iris.sample(frac = 0.75) # 75퍼센트 랜덤하게 뽑음
# 얘를 트레이닝 셋이라고 하고
```

데이터가 너무 작을 때는 cross validation 사용

결정적 단점  
cross validation은 최종 모델이 아님  
대충의 성능 찾기 위해서 다섯개 쪼개서 모델 만들어서 평균 잡음  
평균으로 모델 만들기 때문에 최종 모델로 활용하기가 어려움  
단지 모델에 어느 정도 정확성이 있는지 검증하는 용도  
cross validation <- 오버피팅 체크 위함  
데이터가 작을 때 어느 정도 성능인지 체크하기 위해서  
우리는 데이터 작은 거 관심 없음  
아까는 150개밖에 없었기 때문에 성능 체크하기 위해서 cross validation 방법을 썼었어야 했음

우리 수업은 모델 셀렉션 관점

```python
from sklearn.model_selection import cross_val_score # cross validation도 삼총사 있음
```

```python
cross_val_score(KNeighborsClassifier(), iris.iloc[:,:-1], iris.iloc[:,-1], cv=10, n_jobs=-1) 
# estimator : 알고리즘을 인스턴스화한 애
# 실무에서는 관례상 cv = 10 함
```
