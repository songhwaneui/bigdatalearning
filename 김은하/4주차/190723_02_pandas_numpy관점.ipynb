왜 시리즈?  
우리가 스칼라와 방향 할 때는 방향 있다.  
어레이는 방향 없다.  
가로 방향이든 세로 방향이든 중요하지 않음. 다 시리즈

```python
# 숫자 말고 문자로도 인덱스를 쓸 수 있음
tips.iloc[3] # 숫자 기반 접근 index location 이 번호대로 가지고 와라
```

slice 된다! 넘파이 기반이니까.

```python
tips.iloc[3:6] # 6행 포함되지 않음.
```

```python
tips.loc[3]
```

```python
# Note that contrary to usual python slices, both the start and the stop are included
tips.loc[3:6] # 6행도 포함된다. <- 차이
```

numpy indexing할 때 , ... making 조건 fancyindexing 이 다섯가지도 썼다

```python
tips.loc[3,'day'] # 3번째의 'day'
```

```python
tips.loc[3:6, 'day']
```

```python
tips.loc[3:6, ['day', 'smoker']] # 싹다 넘파이와 같다
```

```python
tips.loc[3:6, ['smoker', 'tip']] # 순서 바꿔도 가져온다
```

```python
tips.iloc[3:6,:-2] # 숫자니까 comma indexing/slicing 지원
```

```python
tips.iloc[3:6,::-1] # 역순 다 numpy 기반이기 때문에
```

```python
# ValueError: Location based indexing can only have [integer, integer slice (START point is INCLUDED, END point is EXCLUDED), listlike of integers, boolean array] types
tips.iloc[3:6,...]
```

```python
tips[['tip','sex']] # 바뀐 채로 가지고 온다
```

```python
tips[['tip']] # fancyindexing 방식을 쓰면 DF로 가지고옴 not series
```

```python
%time tips.iat[3,5] # 하나만 가져오는 애들 iat은 index 방식으로
# 엄청나게 클 때는 얘가 더 빠를 때도 있음
# 보통 i붙은 애가 더 빠름 <- 숫자 기반이기 때문에
```

```python
%time tips.at[3,'time'] # 하나만 가져오는 애들 at은 이름으로
# 그래도 상황에 따라서 다르게 선택해서 쓴다!
```
