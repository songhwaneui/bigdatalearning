### Structured arrays
이것을 기반으로 판다스를 만듬.

```python
x = np.array([('Rex', 9, 81.0),('Fido', 3, 27.0)], 
             dtype=[('name', 'U10'), ('age','<i4'),('weight','<f4')])
```

```python
x[1]
```

###### namedtuple이랑 비슷

```python
from collections import namedtuple
```

```python
Addresss = namedtuple('Addresss',('name','age','weight'))
```

```python
a1 = Addresss('Rex',27,81.0)
```

```python
dir(a1)
```

count와 index는 튜플에 튀어나옴.

```python
a1[0]
```

```python
x["age"]
```

1. 엑셀은 가상 I/O 기때문에 느림
2. numpy는 메모리에 올리기 때문에 빠름

```python
'name' in dir(x) # 어트리뷰트 방식으로는 접근 못함.
```
